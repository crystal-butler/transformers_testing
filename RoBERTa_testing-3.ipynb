```python
# Tokenize a vocabulary word and put it into a torch tensor.
def tokenize_vocab(word):
    vocab_tokens = tokenizer.encode(word)
    vocab_tensor = torch.tensor([vocab_tokens])
    # Print the index of the test word.
    print(f'{vocab_tokens} is length {len(vocab_tokens)}.')
    for t in vocab_tokens[1:-1]:
        print(tokenizer.decode(t))
    return(vocab_tensor)
```

```python
# When a word is found, run the containing line through a fine-tuned RoBERTa model.
def model_text(line):
    line_tokens = tokenizer.encode(line)
    line_tensor = torch.tensor([line_tokens])
    output = model(line_tensor)
#     print(f'{line_tokens} is length {len(line_tokens)}.')
    for t in line_tokens[1:-1]:
        print(f'{tokenizer.decode(t)} is at index {t}.')
    # Map the token strings to their vocabulary indices.
    return(output)
```
